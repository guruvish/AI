```python
# use group by() function to form groups based on more than one category (use more than one column to perform to splliting)
import pandas as pd
df = pd.read_csv("nba-2.csv")
df
```

```python
# applying groupby function to 
# group the data on team value
gk = df.groupby('Team')
# lets print the first entries
# in all groups formed
gk.first()
```

```python
gk.get_group('Boston Celtics')
```

```python
# first grouping based on "team"
# within each team we are grouping based on "position"
gkk = df.groupby(["Team","Position"])
# printing the first value on group
gkk.first()
```

```python
# create simple data frame 
import pandas as pd
import numpy as np

#creating a data frame
df = pd.DataFrame({'A':['john','boby','mina','peter','nicky'],'B':['Masters','Graduate','Graduate','Masters','Graduate'],
                   'C':[27,23,21,23,24]})
df
```

```python
# simplest pivot table must have a data frame
#and an index/list of index
table = pd.pivot_table(df,index=['A','B'])
table
```

```python
# creates a pivot table data frame
table=pd.pivot_table(df,values = 'A',index = ['B','C'],columns=['B'],aggfunc=np.sum)
table

```

```python
# PANDAS PLOTTING
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

ts = pd.Series(np.random.randn(1100),index = pd.date_range('1/1/2000',periods =1100))
ts = ts.cumsum()#cumilative sum
ts.plot()

plt.show()
```
